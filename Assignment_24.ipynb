Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library that provides high-level interfaces for creating informative statistical graphics. Here are five plots we can create using Seaborn and their uses:

Line Plot (sns.lineplot()) – Used for visualizing trends over time, especially for continuous variables.
Bar Plot (sns.barplot()) – Represents categorical data using bars, showing mean values with confidence intervals.
Box Plot (sns.boxplot()) – Displays the distribution of data and highlights outliers using quartiles and whiskers.
Histogram (sns.histplot()) – Shows the frequency distribution of numerical data.
Pair Plot (sns.pairplot()) – Visualizes relationships between numerical variables in a dataset using scatter plots and histograms.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Line Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event & Region")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# Box Plot for Age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Box Plot of Age by Passenger Class")

# Box Plot for Fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot of Fare by Passenger Class")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram of Price with Cut as Hue
plt.figure(figsize=(10, 5))
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True)
plt.title("Distribution of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Pair Plot
sns.pairplot(iris, hue="species")
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data for heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Plot Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm", linewidths=0.5)
plt.title("Number of Passengers Over the Years")
plt.show()
```
