Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library for data visualization. It is mainly used for creating static, animated, and interactive plots in Python. It provides functionalities to represent data graphically, making it easier to understand patterns, trends, and relationships. The Pyplot module of Matplotlib offers a simple interface for generating various types of plots.

Five types of plots that can be created using Pyplot are:

Line Plot – Used for visualizing trends over time.
Bar Plot – Represents categorical data using bars.
Scatter Plot – Shows relationships between two continuous variables.
Histogram – Displays frequency distributions.
Box Plot – Summarizes data distribution and outliers.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting Scatter Plot
plt.scatter(x, y, color='blue', alpha=0.6)
plt.title("Scatter Plot of Random Data")
plt.xlabel("X-axis values")
plt.ylabel("Y-axis values")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating Subplots
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)  # First plot
plt.plot(x, y1, color='red')
plt.title("Line 1")

plt.subplot(2, 2, 2)  # Second plot
plt.plot(x, y2, color='blue')
plt.title("Line 2")

plt.subplot(2, 2, 3)  # Third plot
plt.plot(x, y3, color='green')
plt.title("Line 3")

plt.subplot(2, 2, 4)  # Fourth plot
plt.plot(x, y4, color='purple')
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating Bar Plot
plt.figure(figsize=(10, 4))

# Vertical Bar Plot
plt.subplot(1, 2, 1)
plt.bar(company, profit, color=['red', 'blue', 'green', 'purple'])
plt.title("Company Profits (Vertical Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit")

# Horizontal Bar Plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color=['red', 'blue', 'green', 'purple'])
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit")
plt.ylabel("Company")

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating Box Plot
plt.boxplot([box1, box2], labels=["Dataset 1", "Dataset 2"])
plt.title("Box Plot of Two Datasets")
plt.ylabel("Values")
plt.show()
```
