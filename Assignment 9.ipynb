The ArithmeticError class is a base class for exceptions that occur during arithmetic operations in Python. It serves as a parent class for a variety of arithmetic-related exceptions. Two common exceptions derived from ArithmeticError are ZeroDivisionError and OverflowError.

```python
try:
    result = 10 / 0  # This will raise a ZeroDivisionError
except ZeroDivisionError as e:
    print(f"An error occurred: {e}")
```

```python
try:
    result = 2 ** 10000  # This will raise an OverflowError
except OverflowError as e:
    print(f"An error occurred: {e}")
```

The LookupError class in Python is a base class for exceptions that occur when you try to access elements in a sequence (like a list or a dictionary) using invalid or nonexistent keys or indices. It is a superclass for exceptions that represent various types of lookup-related errors. Two common exceptions derived from LookupError are KeyError and IndexError.

KeyError:

KeyError is raised when you attempt to access a dictionary using a key that does not exist in the dictionary.

```python
my_dict = {'name': 'Alice', 'age': 30}
try:
    value = my_dict['address']  # This will raise a KeyError
except KeyError as e:
    print(f"An error occurred: {e}")
```

IndexError:

IndexError is raised when you attempt to access a sequence (like a list or a string) using an invalid index that is out of range.

```python
my_list = [10, 20, 30]
try:
    value = my_list[4]  # This will raise an IndexError
except IndexError as e:
    print(f"An error occurred: {e}")
```
